### ***train_set , test_set = train_test_split(dataset, test_size=0.2, random_state=42)***

```python
train_set , test_set = train_test_split(housing, test_size=0.2, random_state=42)
print(f"Rows in test data set: {len(test_set)}\n Rows in train set: {len(train_set)}")
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(housing, housing['CHAS']):
  strat_train_set = housing.iloc[train_index]
  strat_test_set = housing.iloc[test_index]
```

```python
strat_test_set['CHAS'].value_counts()     #91/7 =13.0
```

```python
strat_train_set['CHAS'].value_counts()   #361/27 = 13.37037037037037 using almost same ratio as test_set this is the functioning of StratifiedShuffleSplit
```
